## Checking for missing values

```python
Total = df.isnull().sum().sort_values(ascending=False)

Percent=(df.isnull().sum().sort_values(ascending=False))*100/len(df)

missing_data = pd.concat([Total, Percent], axis = 1, keys = ['Total', 'Percentage of Missing Values'])    

missing_data
```

```python
missing_data[missing_data['Percentage of Missing Values']>0]
```

* The features **'umpire3'** are having more than 90% of missing values.So we can drop them.
* The features **'fielder', 'dismissal_kind', 'player_dismissed'** eventhough having 90% of missing values they are representing wicket status.so we shuld keep them. 
* The features **'city','umpire2', 'umpire1', 'player_of_match', 'winner'** are having less missing values.

```python
df.drop(['umpire3'],axis=1,inplace=True)
```

## Winning Decisions 

### Does teams choosing toss affect winning ?

```python
df_matches['toss_decision'].value_counts().plot(kind='pie', fontsize=14, autopct='%3.1f%%', 
                                               figsize=(10,7),startangle=90,shadow=True,legend=True,cmap='GnBu')

plt.ylabel('Toss Decision',fontsize=12)
plt.show()
```

## How toss decision affects match results?

```python
df_matches['toss_win_game_win'] = np.where((df_matches.toss_winner == df_matches.winner),'Yes','No')

plt.figure(figsize = (10,6))
sns.countplot('toss_win_game_win', data=df_matches, hue = 'toss_decision')

plt.title("How Toss Decision affects match result", fontsize=20)
plt.xlabel("Winning match", fontsize = 15)
plt.ylabel("Count", fontsize = 15)
plt.xticks(size = 15)
plt.yticks(size = 15)
plt.show()
```

**Teams which are choosing to field after winning toss has won most of the times**

## Top 10 players with highest Strike rate over recent seasons 

```python
strike_rate = pd.DataFrame({'balls':df.groupby('batsman')['ball'].count(),
                            'run':df.groupby('batsman')['batsman_runs'].sum(),
                            'season':df.groupby('batsman')['season'].nunique()})

strike_rate['strike_rate'] = strike_rate['run']/strike_rate['balls']*100

strike_rate.reset_index(inplace = True)

strike_rate[strike_rate.season.isin([10,11])].sort_values(by = 'strike_rate',ascending = False).head(10)
```
